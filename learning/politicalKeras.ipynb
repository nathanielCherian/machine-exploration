```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import RMSprop

feature_names =  ['party','handicapped-infants', 'water-project-cost-sharing', 
                    'adoption-of-the-budget-resolution', 'physician-fee-freeze',
                    'el-salvador-aid', 'religious-groups-in-schools',
                    'anti-satellite-test-ban', 'aid-to-nicaraguan-contras',
                    'mx-missle', 'immigration', 'synfuels-corporation-cutback',
                    'education-spending', 'superfund-right-to-sue', 'crime',
                    'duty-free-exports', 'export-administration-act-south-africa']

voting_data = pd.read_csv('MLCourse/house-votes-84.data.txt', na_values=['?'], 
                          names = feature_names)
voting_data.head()
```

```python
voting_data.describe()
```

```python
voting_data.dropna(inplace=True)
voting_data.describe()
```

```python
voting_data.replace(('y','n'), (1, 0), inplace=True)
voting_data.replace(('democrat', 'republican'), (1,0), inplace=True)
voting_data.head()
```

```python
all_features = voting_data[feature_names].drop('party', axis=1).values
all_classes = voting_data['party']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(all_features, all_classes, test_size=0.2, random_state=420)
X_train.shape
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', kernel_initializer='normal', input_dim=16))
model.add(Dense(32, activation='relu', kernel_initializer='normal'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    batch_size=30,
                    epochs=20,
                    verbose=2,
                    validation_data=(X_test, y_test)
                   )
```

```python
def create_model():
    model = Sequential()
    model.add(Dense(64, activation='relu', kernel_initializer='normal', input_dim=16))
    model.add(Dense(32, activation='relu', kernel_initializer='normal'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
    return model
```

```python
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

estimator = KerasClassifier(build_fn=create_model, epochs=100, verbose=0)

scores = cross_val_score(estimator, all_features, all_classes, cv=10)
```
